Step 1 — Import Libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim

print(torch.__version__)
```

Step 2 — Define Model

```python
class SimpleModel(nn.Module):

    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(10, 50)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(50, 2)

    def forward(self, x):

        return self.fc2(
            self.relu(
                self.fc1(x)
            )
        )
```

Step 3 — Train Model

```python
model = SimpleModel()

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)

for epoch in range(5):

    x = torch.randn(32, 10)
    y = torch.randint(0, 2, (32,))

    optimizer.zero_grad()

    output = model(x)

    loss = criterion(output, y)

    loss.backward()

    optimizer.step()

    print("Epoch:", epoch, "Loss:", loss.item())

torch.save(
    model.state_dict(),
    "model_state.pth"
)

print("Model saved")
```

Step 4 — Quantize Model

```python
model = SimpleModel()

model.load_state_dict(
    torch.load("model_state.pth")
)

model.eval()

quantized_model = torch.quantization.quantize_dynamic(
    model,
    {nn.Linear},
    dtype=torch.qint8
)

print("Quantized successfully")
```

Step 5 — Save Quantized Model (Safe Method)

```python
torch.save(
    quantized_model.state_dict(),
    "quantized_state.pth"
)

print("Quantized state saved")
```

This avoids:

UnpicklingError
Security issues
Version mismatch
Deployment failures

Step 6 — Load Quantized Model (Correct Way)

```python
model = SimpleModel()

quantized_model = torch.quantization.quantize_dynamic(
    model,
    {nn.Linear},
    dtype=torch.qint8
)

quantized_model.load_state_dict(
    torch.load(
        "quantized_state.pth",
        map_location="cpu"
    )
)

quantized_model.eval()

print("Quantized model loaded successfully")
```

Step 7 — Test Inference

```python
sample = torch.randn(1, 10)

with torch.no_grad():

    output = quantized_model(sample)

prediction = torch.argmax(output)

print("Prediction:", prediction.item())
```

Step 8 — Compare Model Size

```python
import os

original = os.path.getsize("model_state.pth")
quantized = os.path.getsize("quantized_state.pth")

print("Original KB:", original / 1024)
print("Quantized KB:", quantized / 1024)
```
